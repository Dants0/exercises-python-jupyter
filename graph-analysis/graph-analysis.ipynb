```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')
data.replace('?', np.nan, inplace= True)
data = data.astype({'age': np.float64, 'money': np.float64})

fig, axs = plt.subplots(ncols=2, figsize=(20, 4.8))
sns.boxplot(x='age', y='firstname', data=data, ax=axs[0])
sns.violinplot(x='firstname', y='money', data=data, ax=axs[1])
```
